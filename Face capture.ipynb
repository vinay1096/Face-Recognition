```python
import cv2
import numpy as np

# har cascade frontal face detection
face_classifier=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")


def face_extraction(img):
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    if faces is ():
        return None

# Cropping all the faces fopund in the frame
    for (x,y,w,h) in faces:
        x=x-10
        y=y-10
        cropped_face = img[y:y+h+50, x:x+w+50]
        return cropped_face

# initialize webcam
cap = cv2.VideoCapture(0)
count=0
```

```python
while True:
    ret,frame=cap.read()
    if face_extraction(frame) is not None:
        count+=1
        face = cv2.resize(face_extraction(frame), (400,400))
        face=cv2.cvtColor(face,cv2.COLOR_BGR2GRAY)
      
        
        file_name_path="./Samples/" + str(count) + '.jpg'
        cv2.imwrite(file_name_path, face)
        cv2.putText(face,str(count),(50,50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
        cv2.imshow("crop",face)
    else:
        print("No face found")
        pass
    if cv2.waitKey(1) == 13 or count == 10:
        break

cap.release()
cv2.destroyAllWindows()
print("collecting samples")
```
